```python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display, clear_output

# ---------------------------------------------------
# STEP 1: Create dataset inside the code
# ---------------------------------------------------

data = {
    "Movie Name": [
        "Avengers","John Wick","Fast and Furious","The Dark Knight","Mission Impossible",
        "Mad Max Fury Road","Gladiator","Black Panther","Spider Man Homecoming","Iron Man",
        "The Hangover","Mr Bean Movie","Home Alone","Rush Hour","Jumanji",
        "Deadpool","The Mask","Dumb and Dumber","Bruce Almighty","Johnny English",
        "Titanic","The Notebook","La La Land","Me Before You","Pride and Prejudice",
        "The Fault in Our Stars","Notting Hill","Before Sunrise","Love Actually","A Walk to Remember",
        "The Conjuring","Insidious","Annabelle","The Nun","A Quiet Place",
        "IT","The Ring","The Exorcist","Lights Out","The Grudge",
        "Inception","Shutter Island","Gone Girl","The Prestige","Se7en",
        "Fight Club","The Sixth Sense","Parasite","Zodiac","The Girl with the Dragon Tattoo"
    ],

    "Genre": [
        "Action","Action","Action","Action","Action",
        "Action","Action","Action","Action","Action",
        "Comedy","Comedy","Comedy","Comedy","Comedy",
        "Comedy","Comedy","Comedy","Comedy","Comedy",
        "Romance","Romance","Romance","Romance","Romance",
        "Romance","Romance","Romance","Romance","Romance",
        "Horror","Horror","Horror","Horror","Horror",
        "Horror","Horror","Horror","Horror","Horror",
        "Thriller","Thriller","Thriller","Thriller","Thriller",
        "Thriller","Thriller","Thriller","Thriller","Thriller"
    ]
}

df = pd.DataFrame(data)

# Save CSV automatically
df.to_csv("movies.csv", index=False)

# ---------------------------------------------------
# STEP 2: GUI Setup
# ---------------------------------------------------

title = widgets.HTML("<h2 style='color:blue;'>Movie Recommendation System 🎬</h2>")

genre_dropdown = widgets.Dropdown(
    options=sorted(df['Genre'].unique()),
    description='Genre:',
)

button = widgets.Button(
    description="Recommend Movies",
    button_style='success'
)

output = widgets.Output()

# ---------------------------------------------------
# STEP 3: Recommendation Function
# ---------------------------------------------------

def recommend_movies(b):
    with output:
        clear_output()

        selected_genre = genre_dropdown.value
        recommend = df[df['Genre'] == selected_genre]

        print("Top Recommended Movies:\n")

        for i, movie in enumerate(recommend['Movie Name'], start=1):
            print(i, ".", movie)

# Button click event
button.on_click(recommend_movies)

# ---------------------------------------------------
# STEP 4: Display GUI
# ---------------------------------------------------

display(title)
display(genre_dropdown)
display(button)
display(output)
```
